```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
import pickle
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
df = pd.read_csv("laptop_price.csv",encoding = "ISO-8859-1")
```

```python
df.head()
```

```python
df.drop(columns=['OpSys','TypeName','Product'],inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df.drop(columns=['laptop_ID'],inplace = True)
```

```python
df.head()
```

```python
df.shape
```

```python
df['Company'].unique()
```

```python
df.drop(columns=['ScreenResolution'],inplace=True)
```

```python
df.head()
```

```python
df['Memory'].unique()
```

```python
df['Weight'] = df['Weight'].str.replace("kg","")
```

```python
df.head()
```

```python
df['Memory'] = df['Memory'].str.replace("Flash Storage","SSD")
```

```python
df.head()
```

```python
df.shape
```

```python
df.isna().sum()
```

```python
cp = LabelEncoder()
```

```python
df['CPU'] = cp.fit_transform(df['Cpu'])
```

```python
df.head()
```

```python
com = LabelEncoder()
```

```python
df['COMPANY'] = com.fit_transform(df['Company'])
```

```python
df.head()
```

```python
mem = LabelEncoder()
df['MEMORY'] = mem.fit_transform(df['Memory'])
```

```python
df.head()
```

```python
df['Ram'] = df['Ram'].str.replace("GB","")
```

```python
df.head()
```

```python
gp = LabelEncoder()
df['GPU'] = gp.fit_transform(df['Gpu'])
```

```python
df.head()
```

```python
df.to_csv('encoded.csv')
```

```python
spec = df.drop(columns=['Company','Cpu','Memory','Price_euros','Gpu'])
```

```python
spec
```

```python
price = df['Price_euros']
```

```python
lr = LinearRegression()
lrl = Lasso()
lrr = Ridge()
dtr = DecisionTreeRegressor()
rfr = RandomForestRegressor()
```

```python
sc = StandardScaler()
po = PolynomialFeatures(degree=5)
spec_sc = sc.fit_transform(spec)
spec_poly = po.fit_transform(spec)
X_train,X_test,y_train,y_test = train_test_split(spec,price,test_size=0.25)
```

```python
lr.fit(X_train,y_train)
lrl.fit(X_train,y_train)
lrr.fit(X_train,y_train)
dtr.fit(X_train,y_train)
rfr.fit(X_train,y_train)
```

```python
a = lr.score(X_test,y_test)
b = lrl.score(X_test,y_test)
c = lrr.score(X_test,y_test)
d = dtr.score(X_test,y_test)
e = rfr.score(X_test,y_test)
```

```python
print("Linear Regression", a*100)
print("Lasso ",b*100)
print("Ridge ",c*100)
print("Decision tree regressor ",d*100)
print("Random Forest Regressor ",e*100)
```

```python
y_pred = rfr.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
pickle.dump(rfr,open("laptop_pred.pkl","wb"))
```
